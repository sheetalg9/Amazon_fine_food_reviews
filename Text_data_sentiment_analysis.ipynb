```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

```

```python
def preprocessing(raw_text):
    
    txt = re.sub('[^a-zA-Z]' ," ",raw_text)  # removing special charcters
    
    lower = txt.lower() # converting to lower case
    
    split = lower.split()
    
    stp_wrd = [i for i in split if not i in stopwords.words('english')]  # removing stopwords
    
    lemma = WordNetLemmatizer()
    clean_ = [lemma.lemmatize(i) for i in stp_wrd ]  # lemmitization
    
    return pd.Series([" ".join(clean_)])
```
